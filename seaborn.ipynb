```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

sns.lineplot(x=x, y=y, marker='o')
plt.title('PLOTTING POINTS')
plt.show()
```

```python
#seaborn me plot ke liye dataframe bnana hi pdta h 
data = {
        'AGE' : [10,20,30,40],
        'MARKS' : [95,96,92,91]
       }
df = pd.DataFrame(data)
```

```python
sns.lineplot(x= 'AGE',y= 'MARKS',data=df)
plt.show()  ## baaki sb chiz ke liye extra features graph me dikhane ke liye matplotlib use krna pdta h
```

### SCATTER PLOT


```python
sns.scatterplot(x='AGE', y='MARKS', data=df, s=100)
plt.title("AGE vs MARKS")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# 1st subplot
plt.subplot(1, 2, 1)   # 1 row, 2 columns, subplot 1
sns.lineplot(x=x, y=y, marker='o')
plt.title('FIRST PLOT')

# 2nd subplot
plt.subplot(1, 2, 2)   # 1 row, 2 columns, subplot 2
sns.barplot(x=x, y=y)
plt.title('SECOND PLOT')

plt.tight_layout()  # overlap avoid karne ke liye
plt.show()




```

### DIStPLOT


Displot helps to do a univariate analysis on a continuous numerical data.

Displot contains 3 things in one graph

1. Histogram
2. Kernel density Estimation(PDF)
3. Rugplot

```python
# for only histogram make kde = false
## for pdf make kde = true
```

### HISTOGRAM PLOTTING ONLY

```python
# Example dataset
tips = sns.load_dataset("tips")   # built-in dataset with restaurant bills

#make kde = false 
x = tips['total_bill']
sns.displot(x, kde= False,bins = 20)   #bins
plt.show()
```

### Probability Density Function 

```python
### for PDF or kde make kind = kde
x = tips['total_bill']
   #bins

sns.displot(x, kind="kde")
plt.show()
```

### COUNTPLOT

```python
#count plot used in categorical column plotting to show frequency of various categories

data = {
    'Name': ['Aman','Mohan','Rohan','Komal','Radha','Sita'],
    'Sex' : ['Male', 'Male' , 'Male' ,'Female', 'Female', 'Female']
    }
df = pd.DataFrame(data)
sns.countplot(x='Sex', data=df)
plt.show()
```
