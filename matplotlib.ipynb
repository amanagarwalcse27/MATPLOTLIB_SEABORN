```python
import matplotlib.pyplot as plt
import numpy as np
```

### LINE PLOT

```python
x = ['PYTHON','JAVA','C','C++']
y = [100,75,125,25]
plt.plot(x,y)
plt.show()
```

```python
x = ['PYTHON','JAVA','C','C++']
y = [100,75,125,25]
plt.plot(x,y)
plt.title("POPULARITY")

plt.xlabel("LANGUAGE")
plt.ylabel("NO OFSTUDENTS")
```

```python
x = ['PYTHON','JAVA','C','C++']
y = [100,75,125  ,25]
plt.plot(x,y,linestyle = '--',color= 'Red',marker= 'o',markerfacecolor = 'black')
plt.title("POPULARITY")
plt.grid()

plt.xlabel("LANGUAGE")
plt.ylabel("NO OFSTUDENTS")
```

### BAR PLOT

```python

x = ['PYTHON','JAVA','C','C++']
y = [100,75,50,25]



plt.xlabel("LANGUAGE")
plt.ylabel("POPULARITY")
plt.title("CODING",fontsize = 20)
plt.bar(x,y)
plt.show()
```

```python
x = ['PYTHON','JAVA','C','C++']
y = [100,75,50,25]



plt.xlabel("LANGUAGE")
plt.ylabel("POPULARITY ")
plt.title("LANGUAGES POPULARITY COMPARISON",fontsize = 20)

c = ["y","b","m","g"]
plt.bar(x,y,width=0.4,color = c,label = " VARIOUS LANGUAGES")
plt.legend()

plt.show()
```

```python
x = ['PYTHON','JAVA','C','C++']
y = [100,75,50,25]
z = [20,30,40,50]
width = 0.2
p = np.arange(len(x))
p1= [j+width for j in p]




plt.xlabel("LANGUAGE")
plt.ylabel("POPULARITY")
plt.title("CODING",fontsize = 20)

plt.bar(p,y,width,color = 'r',label = " POPULARITY")
plt.bar(p1,z,width,color='y',label = 'popularity1')

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

### PIE CHART

```python
regions = ['North','South','East','West']
revenue = [3000,2000,1000,1500]

plt.pie(revenue, labels= regions,autopct='%.1f%%',colors= ['red','blue','green','orange'])
plt.title('Contribution by Each Region')
plt.show()
```

### histogram
#### used for distribution of data in ranges

```python
scores = [40,50,55,67,80,85,70,40,30,75,68,42]

plt.hist(scores, bins= 5,color = 'purple',edgecolor = 'black')
plt.xlabel('SCORE RANGE')
plt.ylabel('NO OF STUDENTS')

plt.title('SCORE DISTRIBUTION OF STUDENTS')
plt.show()
```

### SCATTER PLOT


```python
x = [2,3,4,5]
y = [30,35,40,44]

plt.xlabel('No of Hours Studied')
plt.ylabel('MARKS OBTAINED')
plt.title('STUDY HOURS')

plt.grid()
plt.scatter(x,y,marker= 'o')
```

### PLOTTING SUBPLOTS

```python
x = [1,2,3,4]
y = [10,20,30,40]

plt.subplot(1,2,1) #1 row, 2 column, 1 subplot
plt.title('FIRST PLOT')
plt.plot(x,y)

plt.subplot(1,2,2) #1 row,2 column, 2 subplot
plt.bar(x,y)
plt.title('SECOND PLOT')
```
